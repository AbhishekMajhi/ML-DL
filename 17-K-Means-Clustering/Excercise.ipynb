```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
# As we know that 
```

```python
# Fitting training data
kmeans.fit(data[0])
```

```python
# Here what KMeans predicted what the cluster will be
# Here the  vlues will be defer in between 0-3
kmeans.labels_
```

 Let' s plot out the difference between KMeans prediction of clustering and orignal clusters

```python
fig,(ax1,ax2) = plt.subplots(1,2, sharey  = True, figsize = (10,7))

ax1.set_title("K Means")
ax1.scatter(data[0][:,0],data[0][:,1], c = kmeans.labels_, cmap ='rainbow')

ax2.set_title("Orignal clusters")
ax2.scatter(data[0][:,0],data[0][:,1], c = data[1], cmap='rainbow')
```
